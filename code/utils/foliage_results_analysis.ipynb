```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles

df = pd.read_csv(r"C:\Users\Hashim\GitHub\DigitalTwin\code\ultralytics\examples\datasets\Grid_SubSample_1_DTRF_LiDAR_csv.csv")


# Filter data for campaigns A and B
campaign_a = df[df['DTRF_1'] == 1]
campaign_b = df[df['LIDAR_1'] == 1]
campaign_a_or_b = df[(df['LIDAR_1']==1) | (df['DTRF_1']==1)]
campaign_a_and_b = df[(df['LIDAR_1']==1) & (df['DTRF_1']==1)]

# Create sets from client IDs
dtrf = set(campaign_a['id'])
lidar = set(campaign_b['id'])
dtrf_or_lidar = set(campaign_a_or_b['id'])
dtrf_and_lidar = set(campaign_a_and_b['id'])

print("DTRF = {}\nLiDAR = {}\nDTRF OR LiDAR = {}\nDTRF AND LiDAR = {}".format(len(dtrf),len(lidar), len(dtrf_or_lidar), len(dtrf_and_lidar)))
print("\n\nOut of the LiDAR reprsented vegetation, DTRF is representing = {:.4f}%".format(len(dtrf_and_lidar)/len(lidar) * 100))
# Plot the Venn diagram
plt.figure(figsize=(8, 8))
# venn2([dtrf, lidar], ('DTRF', 'LiDAR'))


vc=venn2_circles(subsets= [dtrf, lidar],linestyle="dashed",linewidth=2)
vc[0].set_lw(8.0)
vc[0].set_ls('dashed')
vc[0].set_color('limegreen')
vc[0].set_alpha(0.7)

vc[1].set_lw(8.0)
vc[1].set_ls('dashed')
vc[1].set_color('yellow')
vc[1].set_alpha(0.7)

v1 = venn2(
    subsets= [dtrf, lidar],
    set_labels=['DTRF', 'LiDAR'],
    set_colors=['limegreen', 'yellow'], alpha=0.7,
    #subset_label_formatter=lambda x: f"{(x/len(dtrf_lidar)):1.0%}"
    subset_label_formatter=lambda x: str(x) + "\n(" + f"{(x/len(dtrf_or_lidar)):1.0%}" + ")"
)

v1.get_label_by_id('10').set_text('Foliage from \nDT\n{:.2f}%'.format(len(dtrf-lidar)/len(dtrf_or_lidar) * 100))
v1.get_label_by_id('01').set_text('Folaige from \nLiDAR\n{:.2f}%'.format(len(lidar-dtrf)/len(dtrf_or_lidar) * 100))
v1.get_label_by_id('11').set_text('Folaige from \nDT & LiDAR = {:.2f}%\n\n\n Out of LiDAR, \nDTRF represents\n\n{:.2f}%  foliage'.format((len(dtrf&lidar)/len(dtrf_or_lidar) * 100),len(dtrf_and_lidar)/len(lidar) * 100))
v1.get_patch_by_id("11").set_color("green")

# title of the venn diagram 
plt.title("\nFoliage Comparision measured from DT vs LiDAR\n")

plt.savefig(r"C:\Users\Hashim\GitHub\DigitalTwin\code\ultralytics\examples\datasets\Grid_SubSample_1_DTRF_LiDAR_results.png")
plt.show()
```
